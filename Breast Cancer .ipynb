 **PCA is a transformation of data and it helps to find out which featues explain the most variance in your data.**

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data= scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
df_camp=pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_camp,cmap='copper')
```

**As you can see the heatmap shows the correlation of the various features and the principal component themselves. Basically each principal component is shown as row and higher the number lighter the color gets(depends upon the color)  ie its more correlated to a specific feature in the columns.**
